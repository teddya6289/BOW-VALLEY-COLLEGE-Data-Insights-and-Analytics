```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

path = os.path.abspath('C:/Users/Balli/Pictures/prediction')
bb_customer_ratings =pd.read_csv(path +'/bb-customer-ratings.csv')
bb_customer_ratings.info()

# Features for clustering
features = bb_customer_ratings[['customer_retention_score','customer_profitability_score']]

scaler = StandardScaler()

scaled_feature = scaler.fit_transform(features)


# Choosing the number of clusters
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(scaled_feature)
bb_customer_ratings['Cluster'] = kmeans.labels_
# THIS MODEL USES NUMBER FROM 0 TO CLUSTER DATA BASED A SIMILAR BEHAVIOR OF THE DATA
# NOTE: 0 INDICATES LOYAL CUSTOMER GROUP. 1 INDICATES LESS LOYAL CUSTOMER GROUP

loyal_profitable_cust = bb_customer_ratings[(bb_customer_ratings['Cluster']== 0)&(bb_customer_ratings['customer_profitability_score']>0)]
loyal_less_profitable_cust = bb_customer_ratings[(bb_customer_ratings['Cluster']== 0)&(bb_customer_ratings['customer_profitability_score']<=0)]

less_loyal_profitable_cust = bb_customer_ratings[(bb_customer_ratings['Cluster']== 1)&(bb_customer_ratings['customer_profitability_score'] >0)]
less_loyal_less_profitable_cust = bb_customer_ratings[(bb_customer_ratings['Cluster']== 1)&(bb_customer_ratings['customer_profitability_score'] <=0)]

# Group by cluster to understand the segment characteristics
cluster_summary = bb_customer_ratings.groupby('Cluster').mean()


print(cluster_summary)
print(f'LOYAL AND PROFITABLE CUSTOMER GROUP\n{loyal_profitable_cust}')
print(f'LOYAL AND LESS PROFIABLE CUSTOMER GROUP\n{loyal_less_profitable_cust}')
print(f'LESS LOYAL AND PROFITABLE CUSTOMER GROUP\n{less_loyal_profitable_cust}')
print(f'LESS LOYAL AND LESS PROFITABLE CUSTOMER GROUP\n{less_loyal_less_profitable_cust}')

```
